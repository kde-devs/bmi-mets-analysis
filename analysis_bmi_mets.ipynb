# 2019

```python
import pandas as pd

df_2019 = pd.read_csv('/content/drive/MyDrive/공모전/대한임상학회/KNHANES_2019_BMI_MET.csv', encoding='MacRoman', on_bad_lines='skip')
print(df_2019)
```

```python
# 필요한 변수 목록
columns_to_extract = [
    "HE_BMI", "HE_obe", "HE_wc",
    "HE_sbp", "HE_dbp", "HE_HP",
    "HE_glu", "HE_HbA1c", "HE_DM_HbA1c",
    "HE_chol", "HE_HDL_st2", "HE_TG", "HE_LDL_drct",
    "HE_HCHOL", "HE_HTG",
    "sex", "age", "incm"
]

# 변수 추출
df_2019_selected = df_2019[columns_to_extract]


df_2019_selected
```

```python
# 저장 (선택)
df_2019_selected.to_csv("BMI_대사질환_데이터_2019.csv", index=False)
```

```python
# Count the number of NaN values in each column
nan_counts = df_2019_selected.isnull().sum()

# Display the counts
print("Number of NaN values per column:")
print(nan_counts)
```
